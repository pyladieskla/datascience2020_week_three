## 1.compute the mean, standard deviation, and variance of the given array along the second axis using numpy

```python
data = np.array([[11, 22],[33, 44],[55, 66]])
```

```python
#Mean using numpy.mean() inbuilt function
m1 = np.mean(data)
print("Mean with inbuilt method: ", m1)

#Mean using a formula
m2 = np.average(data)
print("Mean with a formula: ", m2)
```

```python
#Standard deviation using numpy.std() inbuilt function
s1 = np.std(data)
print("Standard deviation with inbuilt method: ", s1)

#Standard deviation using a formula
s2 = np.sqrt(np.mean((data - np.mean(data)) ** 2))
print("Standard deviation with a formula: ", s2)
```

```python
#Variance using numpy.var() inbuilt function
v1 = np.var(data)
print("Variance with inbuilt method: ", v1)

#Variance using a formula
v2 = np.mean((data - np.mean(data)) ** 2)
print("Variance with a formula: ", v2)
```

### 4. Read the titanic dataset into a dataframe

### 6.What is the relationship between age and fare

```python
#Relationship between age and fare using corr(), corelationship
column_1 = titanic_df['age']
column_2 = titanic_df['fare']
relationship = column_1.corr(column_2)
print(relationship)
```

### 7. Calculate the probability of survival

```python
#Probability of survival
#Using .groupby() and the inbuilt .div()
survival_probs = titanic_df.groupby('survived').size().div(len(titanic_df))
print(survival_probs)
```
