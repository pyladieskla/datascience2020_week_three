## 1.compute the mean, standard deviation, and variance of the given array along the second axis using numpy

```python
data = np.array([[11, 22],[33, 44],[55, 66]])
```

```python
data_second_row = data[1]

mean, std, variance = np.mean(data), np.std(data), np.var(data)
mean_second_row, std_second_row, variance_second_row = np.mean(data_second_row), np.std(data_second_row), np.var(data_second_row)

title = "Mean  Standard-Deviation  Variance".rjust(47)
formatted_title = "\033[1m" + title + "\033[0m"
space = "".rjust(15)

print("{} \nWhole array  {}  {}  {} \nSecond row   {}  {} {} {}".format(formatted_title, mean, std, variance, mean_second_row, std_second_row, space, variance_second_row))
```
